# 1 Importing Packages

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

# 2 Data

## 2.1 Load Data

```python
data = pd.read_csv('C:/Users/LucasFranklin/Documents/Data Mining/b5.csv')
```

```python
data
```

## 2.2 Convert data to numpy array

```python
X = data.values
```

### 2.2.1 Scale the values  

```python
X = scale(X)
```

# 3 Analysis
## PCA with 50 components

```python
pca = PCA()
```

```python
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_)
```

## PCA with 5 components 

```python
pca5 = PCA(n_components = 5)
pca5.fit(X)
```

```python
print(pca5.explained_variance_ratio_)
```

## Contrinuitions of variables to components 

```python
print(pca5.components_)
```
